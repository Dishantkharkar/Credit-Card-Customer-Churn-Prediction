```python
import tensorflow

from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Conv2D, Flatten
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# train the model with the padding = 'valid'

model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
# train the model with paddding = 'same'

model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
# train the model with paddding = 'same' and strides=(2,2)

model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

# Model with padding 'valid'
model_valid = Sequential()
model_valid.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu', input_shape=(28,28,1)))
model_valid.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'))
model_valid.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'))
model_valid.add(Flatten())
model_valid.add(Dense(128, activation='relu'))
model_valid.add(Dense(10, activation='softmax'))

# Save the model summary to a file
with open('model_valid_summary.txt', 'w') as f:
    model_valid.summary(print_fn=lambda x: f.write(x + '\n'))

# Plot the model architecture
plot_model(model_valid, to_file='model_valid_plot.png', show_shapes=True, show_layer_names=True)
```
